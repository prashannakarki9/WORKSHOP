```python
# Subset for passengers in second class (Pclass is 2)
second_class_fare = data[data['Pclass'] == 2]['Fare'].sum()

# Subset for passengers in third class (Pclass is 3)
third_class_fare = data[data['Pclass'] == 3]['Fare'].sum()

second_class_fare, third_class_fare
```

```python
# Create a new column 'age_group' that categorizes passengers based on age
data['age_group'] = pd.cut(data['Age'], bins=[0, 18, 64, 100], labels=['child', 'adult', 'senior'], right=False)

# Display the updated dataset with the new 'age_group' column
data[['Name', 'Age', 'age_group']].head()
```

```python
# Calculate the total number of passengers
total_passengers = len(data)
total_passengers
```

```python
# Count the number of passengers in each age group
age_group_counts = data['age_group'].value_counts()
age_group_counts
```

```python
# Calculate the proportion of passengers in each age group
age_group_proportions = age_group_counts / total_passengers
age_group_proportions
```

```python
# Display the proportion as a percentage
age_group_percentage = age_group_proportions * 100
age_group_percentage
```
